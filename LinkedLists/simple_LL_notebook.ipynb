```python
from typing import Optional, List
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def build_linked_list(values: List[int]) -> Optional[ListNode]:
    if not values:
        return None
    head=ListNode(values[0])
    current=head
    for val in values[1:]:
        current.next=ListNode(val)
        current=current.next
    return head
```

```python
def print_linked_list(head):
    if not head:
        return None
    result=[]
    curr=head
    while curr:
        result.append(str(curr.val))
        curr=curr.next
    print('->'.join(result))
```

```python
def insert_at_position(head: ListNode, position: int, value: int) -> ListNode:
    # Create the new node to be inserted
    new_node = ListNode(val=value)

    # --- Case 1: Inserting at the head (position 0) ---
    # This also handles inserting into an empty list (where head is None)
    if position == 0:
        new_node.next = head
        return new_node

    # --- Case 2: Inserting in the middle or at the end ---
    current = head
    current_position = 0

    # Traverse to the node *just before* the target position
    # The loop stops when 'current' is the node at (position - 1)
    while current_position < position - 1 and current is not None:
        current = current.next
        current_position += 1

    # If 'current' is None, the position is out of bounds (e.g., trying to
    # insert at position 10 in a list of length 5). We can just return the
    # original head without making changes.
    if current is None:
        return head

    # Perform the insertion
    new_node.next = current.next
    current.next = new_node

    return head
```

```python
def find_middle_linked_list(head):
    if not head:
        return None
    slow=head
    fast=head
    while fast and fast.next:
        slow=slow.next
        fast=fast.next.next
    return slow
```

```python
def reverse_linked_list(head):
    if not head:
        return None
    previous=None
    current=head
    while current:
        aux_temp=current.next
        current.next=previous
        previous=current
        current=aux_temp
    head=previous
    return head
```
